```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# XOR inputs
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]])

# XOR outputs (labels)
y = np.array([0, 1, 1, 0])  # XOR truth table
```

```python
X_bias = np.c_[np.ones((X.shape[0], 1)), X]  # add bias term (x0 = 1)
```

```python
np.random.seed(42)
weights = np.random.randn(3)  # 2 inputs + bias
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
learning_rate = 0.1
epochs = 20

for epoch in range(epochs):
    total_error = 0
    for i in range(X_bias.shape[0]):
        z = np.dot(X_bias[i], weights)
        y_pred = step(z)
        error = y[i] - y_pred
        weights += learning_rate * error * X_bias[i]
        total_error += abs(error)
    print(f"Epoch {epoch+1} Error: {total_error}")
```

```python
# Plotting
x1 = np.linspace(-0.5, 1.5, 100)
x2 = -(weights[0] + weights[1]*x1) / weights[2]

plt.figure(figsize=(8,6))
plt.plot(x1, x2, label="Decision Boundary")

# Plot points
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red', label='0' if i==0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', label='1' if i==1 else "")

plt.title("Single-Layer Perceptron - XOR (Fails)")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.legend()
plt.grid(True)
plt.show()
```
